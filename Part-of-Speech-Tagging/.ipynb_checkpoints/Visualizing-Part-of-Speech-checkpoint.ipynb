# Visualizing Part of Speech


```python
import spacy
from spacy import displacy
import en_core_web_sm

nlp=en_core_web_sm.load()
```

```python
doc=nlp(u"The quick brown fox jumps over the lazy dog's back")
```

```python
displacy.render(doc,style='dep',jupyter=True,options={'distance':100})
```

```python
options={'color':'black','bg':'#09a3d5','compact':True,'font':'times','distance':100}
displacy.render(doc,style='dep',jupyter=True,options=options)
```

```python
for token in doc:
    print(f"{token.text:{10}} {token.pos_:{10}} {token.dep_:{10}} {spacy.explain(token.dep_)}")
```

### Creating Visualization of pos outside jupyter notebook

```python
displacy.serve(doc,style='dep',options=options)
```
